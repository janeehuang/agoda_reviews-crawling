```python
#ÂºïÂÖ•requests„ÄÅBeautifulSoupÊ®°ÁµÑ
import requests
import pandas as pd
from bs4 import BeautifulSoup
#ÂÆöÁæ©Ë¶ÅÁà¨ÂèñÁöÑURL
url = 'https://www.agoda.com/zh-tw/jingan-classic-inn/reviews/taipei-tw.html?checkIn=2020-03-30&los=1&adults=2&rooms=1&cid=1826371&tag=9036467c-2252-b5c0-dad5-2ddffd49e3fe&searchrequestid=e4aa7e18-b2e8-40c8-883d-3d2fbf38aac5&travellerType=1&tspTypes=2%2C7'
```

```python
#ÂæûurlÁç≤ÂèñË´ãÊ±Ç‰∏¶ÈÄöÈÅéBeautifulSoupÈÄ≤Ë°åËß£Êûê
resp = requests.get(url)
resp.encoding = 'utf-8' # ‰ª•utf-8Ê†ºÂºèÁ∑®Á¢º
soup = BeautifulSoup(resp.text, 'lxml')
```

# ‰ΩøÁî®SeleniumÈÄ≤Ë°åËá™ÂãïÂåñÁà¨ÂèñÊï∏Êìö

```python
from selenium import webdriver
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#‰ΩøÁî®webdriverÊâìÈñãÁÄèË¶ΩÂô®
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get(url)

#‰ΩøÁî®Áõ∏Â∞çË∑ØÂæëÊâæÂá∫element
browser.find_element_by_xpath("//div[@id='reviewFilterSection']/div/div[3]").click()
browser.find_element_by_xpath("//div[@id='reviewFilterSection']/div/div[3]/div/ul/li[2]/span/span[2]").click()
soup = BeautifulSoup(browser.page_source)
#whileËø¥ÂúàÂü∑Ë°åÊòØÂê¶Êúâ„ÄåÊü•ÁúãÊõ¥Â§ö‰ΩèÂÆøË©ïÈëë„ÄçÊåâÈàïÔºåÂ¶ÇÊúâÁπºÁ∫åÁà¨ÂèñË≥áÊñôÔºåÂ¶ÇÁÑ°ÂâáÈÄÄÂá∫‰∏¶ÈóúÈñâ
while len(soup.find_all('button', class_='Review-paginator-button')) >0:
    browser.find_element_by_xpath("//button[contains(.,'Êü•ÁúãÊõ¥Â§ö‰ΩèÂÆøË©ïÈëë')]").click()
    soup = BeautifulSoup(browser.page_source)
soup = BeautifulSoup(browser.page_source)

browser.close()
```

```python
print(soup)
```

# Â∞áÁà¨Âèñ‰∏ã‰æÜÁöÑÔº∏Ôº≠Ôº¨ÈÄ≤Ë°åËß£Êûê‰∏¶Â∞áË©ïË´ñÂÖßÂÆπÂèäË©ïË´ñÂàÜÊï∏ÂàÜÂà•Â≠òÂèñ

```python
reviews_score = []
reviews_body = []
```

```python
#Â≠òÂèñË©ïË´ñÂàÜÊï∏
for ele in soup.find_all('div',class_ = 'Review-comment-leftScore'):
    reviews_score.append([ele.text])
```

```python
#Â≠òÂèñË©ïË´ñÂÖßÂÆπ
for ele in soup.find_all('div',class_ = 'Review-comment-body' ):
    reviews_body.append([ele.text])
```

```python
len(reviews_body)
```

```python
for i in range(len(reviews_body)):
    print(reviews_body[i])
```

```python
reviews = []
```

```python
#Âà©Áî®forËø¥ÂúàÂ∞áÊâÄÊúâË©ïË´ñË≥áÊñôÂèäË©ïË´ñÂàÜÊï∏Êï¥Âêà
for i in range(len(reviews_score)):
    reviews.append(reviews_score[i]+reviews_body[i])
```

```python
reviews
```

## ‰ΩøÁî®PandaÂ∞áË≥áÊñôÂëàÁèæ

```python
df = pd.DataFrame(reviews, columns = ['Score','Reviews'])
```
