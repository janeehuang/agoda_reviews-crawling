# 引入模組及要爬取的ＵＲＬ

```python
#引入requests、BeautifulSoup模組
import requests
import pandas as pd
from bs4 import BeautifulSoup
#定義要爬取的URL
url = 'https://www.agoda.com/zh-tw/jingan-classic-inn/reviews/taipei-tw.html?checkIn=2020-03-30&los=1&adults=2&rooms=1&cid=1826371&tag=9036467c-2252-b5c0-dad5-2ddffd49e3fe&searchrequestid=e4aa7e18-b2e8-40c8-883d-3d2fbf38aac5&travellerType=1&tspTypes=2%2C7'
```

```python
#從url獲取請求並通過BeautifulSoup進行解析
resp = requests.get(url)
resp.encoding = 'utf-8' # 以utf-8格式編碼
soup = BeautifulSoup(resp.text, 'lxml')
```

# 使用Selenium進行自動化爬取數據

```python
from selenium import webdriver
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#使用webdriver打開瀏覽器
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get(url)

#使用相對路徑找出element
browser.find_element_by_xpath("//div[@id='reviewFilterSection']/div/div[3]").click()
browser.find_element_by_xpath("//div[@id='reviewFilterSection']/div/div[3]/div/ul/li[2]/span/span[2]").click()
soup = BeautifulSoup(browser.page_source)
#while迴圈執行是否有「查看更多住宿評鑑」按鈕，如有繼續爬取資料，如無則退出並關閉
while len(soup.find_all('button', class_='Review-paginator-button')) >0:
    browser.find_element_by_xpath("//button[contains(.,'查看更多住宿評鑑')]").click()
    soup = BeautifulSoup(browser.page_source)
soup = BeautifulSoup(browser.page_source)

browser.close()
```

```python
print(soup)
```

# 將爬取下來的ＸＭＬ進行解析並將評論內容及評論分數分別存取

```python
reviews_score = []
reviews_body = []
```

```python
#存取評論分數
for ele in soup.find_all('div',class_ = 'Review-comment-leftScore'):
    reviews_score.append([ele.text])
```

```python
#存取評論內容
for ele in soup.find_all('div',class_ = 'Review-comment-body' ):
    reviews_body.append([ele.text])
```

```python
len(reviews_body)
```

```python
for i in range(len(reviews_body)):
    print(reviews_body[i])
```

```python
reviews = []
```

```python
#利用for迴圈將所有評論資料及評論分數整合
for i in range(len(reviews_score)):
    reviews.append(reviews_score[i]+reviews_body[i])
```

```python
reviews
```

## 使用Panda將資料呈現

```python
df = pd.DataFrame(reviews, columns = ['Score','Reviews'])
```
